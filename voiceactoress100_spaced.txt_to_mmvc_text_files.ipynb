このコードは、[Hiroshiba氏の作成したJVSコーパス](https://github.com/Hiroshiba/jvs_hiho)のvoiceactoress100_spaced.txtを、  
MMVC向けにファイルを分割するものです。

```python
from google.colab import drive
drive.mount('/content/drive')

import os
import urllib.request, urllib.error
from os.path import abspath, join, split

url = 'https://raw.githubusercontent.com/Hiroshiba/jvs_hiho/master/voiceactoress100_spaced.txt'
with urllib.request.urlopen(url) as spaced:
    for i, line in enumerate(spaced.read().decode().split('\n')):
        number = i + 1;

        if number >= 101:
            break

        path = '/content/drive/MyDrive/MMVC_Trainer-main/dataset/textful/000_jvs001/text/VOICEACTRESS100_'+str(number).zfill(3)+'.txt'
        with open(path, mode='w', encoding='utf-8') as fp:
            fp.write(line)
```
